```python
all_laps = []

years = [2022, 2023, 2024]

for year in years:

    print(f"\nLoading {year} season...")

    for rnd in range(1, 23):

        try:
            session = fastf1.get_session(year, rnd, 'R')

            session.load()

            laps = session.laps.copy()

            laps['Year'] = year
            laps['Round'] = rnd

            keep_cols = [
                'Driver',
                'LapNumber',
                'Compound',
                'TyreLife',
                'LapTime',
                'Position',
                'PitInTime',
                'Year',
                'Round'
            ]

            laps = laps[keep_cols]

            all_laps.append(laps)

            print(f"Loaded {year} Round {rnd}")

        except Exception as e:
            print(f"Skipped {year} Round {rnd}: {e}")
```

```python
all_laps = []

race_plan = [
    (2022, 1),
    (2022, 5),
    (2022, 10),

    (2023, 1),
    (2023, 5),
    (2023, 10),

    (2024, 1),
    (2024, 5),
    (2024, 10)
]

for year, rnd in race_plan:

    try:
        print(f"\nLoading {year} Round {rnd}...")

        session = fastf1.get_session(year, rnd, 'R')

        session.load()

        laps = session.laps.copy()

        laps['Year'] = year
        laps['Round'] = rnd

        keep_cols = [
            'Driver',
            'LapNumber',
            'Compound',
            'TyreLife',
            'LapTime',
            'Position',
            'PitInTime',
            'Year',
            'Round'
        ]

        laps = laps[keep_cols]

        all_laps.append(laps)

        print(f"Loaded {year} Round {rnd}")

    except Exception as e:
        print(f"Skipped {year} Round {rnd}: {e}")
```

```python
df = pd.concat(all_laps, ignore_index=True)

print(df.shape)

df.head()
```

```python
df['LapTime_s'] = df['LapTime'].dt.total_seconds()
```

```python
df = df.sort_values(
    ['Year', 'Round', 'Driver', 'LapNumber']
)
```

```python
df['pit_next_lap'] = (
    df.groupby(['Year', 'Round', 'Driver'])['PitInTime']
    .shift(-1)
    .notnull()
    .astype(int)
)
```

```python
df = df.sort_values(['Year', 'Round', 'LapNumber', 'Position'])

df['gap_ahead'] = (
    df.groupby(['Year', 'Round', 'LapNumber'])['LapTime_s']
    .diff()
    .abs()
)
```

```python
df['gap_ahead'] = df['gap_ahead'].fillna(999)
```

```python
df['undercut_risk'] = (
    (df['gap_ahead'] < 22) &
    (df['TyreLife'] > 12)
).astype(int)
```

```python
df['rolling_avg_3'] = (
    df.groupby(['Year', 'Round', 'Driver'])['LapTime_s']
    .transform(lambda x: x.rolling(3).mean())
)

df['deg_rate'] = (
    df['LapTime_s'] -
    df['rolling_avg_3']
)
```

```python
compound_map = {
    'SOFT': 0,
    'MEDIUM': 1,
    'HARD': 2
}

df['CompoundEncoded'] = df['Compound'].map(compound_map)
```

```python
features = [
    'LapNumber',
    'TyreLife',
    'CompoundEncoded',
    'deg_rate',
    'gap_ahead',
    'undercut_risk',
    'Position'
]

df = df.dropna(subset=features)

X = df[features]

y = df['pit_next_lap']

print(X.head())
print(y.value_counts())
```

```python
train_df = df[df['Year'].isin([2022, 2023])]
test_df = df[df['Year'] == 2024]

X_train = train_df[features]
y_train = train_df['pit_next_lap']

X_test = test_df[features]
y_test = test_df['pit_next_lap']

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

model = XGBClassifier(
    objective='binary:logistic',
    n_estimators=200,
    max_depth=5,
    learning_rate=0.05,
    scale_pos_weight=scale,
    random_state=42
)

model.fit(X_train, y_train)

print("Improved Race Brain model trained")
```

```python
def predict_pit(
    lap,
    tyre_life,
    compound,
    deg_rate,
    gap_ahead,
    undercut_risk,
    position
):

    sample = pd.DataFrame([{
        'LapNumber': lap,
        'TyreLife': tyre_life,
        'CompoundEncoded': compound,
        'deg_rate': deg_rate,
        'gap_ahead': gap_ahead,
        'undercut_risk': undercut_risk,
        'Position': position
    }])

    probability = model.predict_proba(sample)[0][1]

    return {
        "probability": round(float(probability), 3),
        "recommended": probability > 0.60,
        "undercut_risk": undercut_risk
    }
```

```python
result = predict_pit(
    lap=28,
    tyre_life=18,
    compound=0,
    deg_rate=2.1,
    gap_ahead=8.5,
    undercut_risk=1,
    position=4
)

print(result)
```
